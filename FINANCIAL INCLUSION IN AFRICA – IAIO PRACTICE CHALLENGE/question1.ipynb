```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
train_data = pd.read_csv('Train.csv')
test_data = pd.read_csv('Test.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
columns = ['country',  'bank_account', 'location_type', 'cellphone_access', 'gender_of_respondent', 
           'relationship_with_head', 'marital_status', 'education_level', 'job_type']
for i in columns:
    if i == 'bank_account':
        train_data[i] = encoder.fit_transform(train_data[i])
    else:
        train_data[i] = encoder.fit_transform(train_data[i])
        test_data[i] = encoder.fit_transform(test_data[i])
```

```python
train_data = train_data.drop(columns=['uniqueid'])
#just remove the id it doesn't contribute at all 
```

```python
#household_size and age_of_respondent should be normalized
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
train_data[['household_size', 'age_of_respondent']] = scaler.fit_transform(train_data[['household_size', 'age_of_respondent']])
```

```python
train_data.describe()
```

```python
train_data.head()
```

```python
#the dataset is ready to be trained
label = train_data['bank_account']
features = train_data.drop(columns=['bank_account'])
```

```python
(features_train, features_test, label_train, label_test) = train_test_split(features, label, test_size=0.7)
```

```python
#trying out different models
model = LogisticRegression(penalty = 'l2', C = 0.1, solver = 'sag')
model.fit(features_train, label_train)
model.score(features_test, label_test)
```

```python
model = DecisionTreeClassifier()
model.fit(features_train, label_train)
model.score(features_test, label_test)
```

```python
model = SVC(C=0.1, kernel = 'rbf', gamma = 0.5)
model.fit(features_train, label_train)
model.score(features_test, label_test)
```

```python
#okay the last score was my best
#now I will perform grid search to find the optimal combination
```

```python
parameters_lr = {
    'penalty': ['l1', 'l2', 'elasticnet', None],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['linear', 'sag']
}
parameters_svc = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'gamma': [0.1, 0.01, 0.001, 0.5],
    'degree': [2, 3, 4, 5],
    'kernel': ['linear', 'rbf', 'polynomial', 'sigmoid']
}
```

```python
from sklearn.model_selection import GridSearchCV
logistic_regressor = LogisticRegression(random_state=42)
optimal_regressor = GridSearchCV(logistic_regressor, parameters_lr, cv=5, scoring = 'accuracy')
optimal_regressor.fit(features_train, label_train)
optimal_regressor.score(features_test, label_test)
```

```python
svc_classifier = SVC(random_state = 42)
optimal_svc = GridSearchCV(svc_classifier, parameters_svc, cv=5, scoring = 'accuracy')
optimal_svc.fit(features_train, label_train)
optimal_svc.score(features_test, label_test)
```

```python
#since the last model produced the best accuracy, we will use that one:
test_data.info()
```
